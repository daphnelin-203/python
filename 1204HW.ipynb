```python
import numpy as np
import matplotlib.pyplot as plt
```

## 利用 Contour 繪製等高線圖

#### 以 0.01 為間隔

```python
x = np.arange(-10,10,0.01)
y = np.arange(-10,10,0.01)
```

```python
X,Y = np.meshgrid(x,y)
```

#### 以函數 z = x**2 + y**2 為例

```python
Z = X**2+Y**2
```

#### 建立畫布

```python
plt.figure(figsize=(10,6))
```

#### 用 contourf 繪製「簡易版」等高線圖形

```python
contour = plt.contour(X,Y,Z,[20,40],colors='k')
plt.clabel(contour,fontsize=10,colors=('k','r'))
```

#### 用 contourf 繪製「複雜版」等高線圖形

##### 1. 定義函數

```python
def height(x, y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
```

##### 2. 建立數據

```python
x = np.linspace(-3,3, 250)
y = np.linspace(-3,3, 250)
```

```python
X,Y = np.meshgrid(x,y)
```

##### 3. 使用 cmap 顏色格式的 hot 熱溫圖
##### 4. 畫上黑色的線標示數值

```python
plt.contourf(X,Y,height(X,Y),8,alpha=0.6, cmap=plt.cm.hot)
C = plt.contour(X,Y,height(X,Y),8, colors='k')
plt.clabel(C,inline=True,fontsize=10)
```
